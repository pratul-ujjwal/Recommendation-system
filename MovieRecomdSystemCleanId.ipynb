```python
import pandas as pd
import numpy as np
import warnings; warnings.simplefilter('ignore')
```

```python
movie_metadata = pd.read_csv('/home/saddie/Documents/CIS5570/movie_dataset/movies_metadata.csv')
movie_metadata_cols = movie_metadata.columns
print(movie_metadata_cols)
print(movie_metadata.dtypes)
movie_metadata.head(2)
# movie_metadata[movie_metadata['id']=='4912']
```

```python
movie_metadata_shape = movie_metadata.shape
print(movie_metadata.shape)
print(movie_metadata['id'].unique().shape)
movie_metadata = movie_metadata.drop_duplicates()
print(movie_metadata.shape)
print(movie_metadata['id'].unique().shape)
```

```python
movie_id_array = movie_metadata['id'].values.astype(str)
movie_id_numeric_bool = np.char.isnumeric(movie_id_array)
movie_id_numeric_array = movie_id_array[movie_id_numeric_bool]
movie_id_numeric_array = movie_id_numeric_array.astype(int)
print(movie_id_numeric_array.shape)
```

```python
movie_metadata.iloc[np.where([movie_id_numeric_bool==False])[1]]
```

```python
temp_index = movie_metadata.iloc[np.where([movie_id_numeric_bool==False])[1]].index
movie_metadata = movie_metadata.drop(temp_index)
```

```python
print(movie_metadata.shape)
```

```python
movie_id_unique_numeric, movie_id_unique_numeric_indices = np.unique(movie_id_numeric_array.astype(int), return_index=True)
print(movie_id_unique_numeric.shape)
movie_id_numeric_not_unique = np.delete(movie_id_numeric_array, movie_id_unique_numeric_indices)
```

```python
movie_id_numeric_not_unique
```

```python
movie_metadata_sorted_id = movie_metadata.take(movie_id_unique_numeric_indices)
movie_metadata_sorted_id.head()
```

```python
movie_metadata.drop(movie_metadata.iloc[movie_id_unique_numeric_indices].index)
```

```python
movie_metadata_sorted_id = movie_metadata_sorted_id.set_index('id')
```
